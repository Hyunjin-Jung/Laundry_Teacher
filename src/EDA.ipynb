```python
import os

label_dir = "./datasets/val/labels/"

labelArr = os.listdir(label_dir)
```

# box 분포

```python
box_counter = dict()
for c in classes:
    box_counter[c] = 0

for labelFile in labelArr:
    file = open(label_dir + labelFile)
    lines = file.readlines()

    class_set = set()
    for line in lines:
        data = line.split(" ")
        class_set.add(int(data[0]))

    for c in class_set:
        box_counter[classes[c]] += 1
    
```

```python
import matplotlib.pyplot as plt

x = list(box_counter.keys())
y = list(box_counter.values())

plt.figure(figsize=(30, 5))
plt.bar(x, y)
plt.xticks(rotation=90, fontsize=10)
plt.show()
```

# fiber 라벨 섞여있는 경우

```python
mixed_label_counter = dict()

for labelFile in labelArr:
    file = open(label_dir + labelFile)
    lines = file.readlines()

    class_set = set()
    for line in lines:
        data = line.split(" ")
        class_set.add(int(data[0])//8)
    
    key = len(class_set)
    if key in mixed_label_counter:
        mixed_label_counter[key] += 1
    else:
        mixed_label_counter[key] = 1
```

# 세탁 방법 분포 

```python
washing_method_classes = [
    "hand washing30",
    "washing30",
    "washing40", #     "washing40_1" : 2,
    "washing60",
    "washing95",

    "dryclean",
    "dryclean : laundry",
    "dryclean : petroleum solvent only"
]
```

```python
washing_method_counter = dict()
for c in washing_method_classes:
    washing_method_counter[c] = 0

for labelFile in labelArr:
    file = open(label_dir + labelFile)
    lines = file.readlines()

    class_set = set()
    for line in lines:
        data = line.split(" ")
        class_set.add(int(data[0])%8)

        for c in class_set:
            washing_method_counter[washing_method_classes[c]] += 1
            break
```

```python
import matplotlib.pyplot as plt

x = list(washing_method_counter.keys())
y = list(washing_method_counter.values())

plt.figure(figsize=(10, 5))
plt.bar(x, y)
plt.xticks(rotation=90, fontsize=10)
plt.show()
```

# 세탁방법 별 fiber 분포 (fiber 1개인 경우)

```python
washing_method_data = list()
for w in washing_method_classes:
    w_arr = [0 for i in range(12)]
    washing_method_data.append(w_arr)
    
for labelFile in labelArr:
    file = open(label_dir + labelFile)
    lines = file.readlines()

    class_set = set()
    for line in lines:
        data = line.split(" ")
        class_set.add(int(data[0]))

        if len(class_set) == 1:
            for c in class_set:
                washing_method = c%8
                fiber = c//8
                washing_method_data[washing_method][fiber] += 1
```

```python
for row in washing_method_data:
    s = sum(row)
    row[:] = [x / s for x in row]
```

```python
import numpy as np
import matplotlib.pyplot as plt

data  = np.array(washing_method_data)

plt.figure(figsize=(10, 5))

x = list(washing_method_classes)
y_last = np.zeros(len(washing_method_classes))
for f_idx in range(len(fiber_classes)):
    y = list(data[:, f_idx])
    if f_idx == 0:
        plt.barh(x, y)
    else:
        plt.barh(x,y, left=y_last)
    y_last += y

plt.xticks(rotation=90, fontsize=10)
plt.show()
```
