### Plotting results.csv resulted from training the model.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./results.csv')
df
```

```python
# Extract fields in the csv
column_names = df.columns.values.tolist()
# column_names = [name.strip() for name in column_names]
x = column_names.pop(0)

# Caculate the number of fields of one type to divide train and val data in a row.
count = 0
for name in column_names:
    if 'val' in name:
        break
    count = count + 1

# Create and draw subplots.
fig, axes = plt.subplots(nrows=2, ncols=count, figsize=(60, 15))
for i, row in enumerate(axes):
    for j, cell in enumerate(row):
        next = i*count + j
        if next < len(column_names):
            cell.plot(df[x], df[column_names[next]], marker="o", mfc='red', mec='k')
            cell.set_title(column_names[next])
fig.savefig('results.png')
```
